### Overview of data 

```python
import pandas as pd
```

```python
# RNC transcripts
import nltk
from nltk.corpus import PlaintextCorpusReader
corpus_root = 'data/RNC/'
reps = PlaintextCorpusReader(corpus_root, '.*txt')
```

```python
# RNC transcripts
import nltk
from nltk.corpus import PlaintextCorpusReader
corpus_root = 'data/DNC/'
dems = PlaintextCorpusReader(corpus_root, '.*txt')
```

```python
transRNC = [reps.raw(x) for x in reps.fileids()]
transDNC = [dems.raw(x) for x in dems.fileids()]
```

```python
DNCspeakeraff = pd.DataFrame({'Speakers': dems.fileids(), 
                       'Aff': 'D',
                       'Speech' : transDNC}) 
RNCspeakeraff = pd.DataFrame({'Speakers': reps.fileids(), 
                       'Aff': 'R',
                        'Speech' :transRNC})
```

```python
speakeraff = pd.concat([DNCspeakeraff,RNCspeakeraff])
speakeraff.set_index('Speakers') #the contents in the data files with their political affiliation
```

```python
len(speakeraff) #There were 51 speakers in total in the two data sets
```

```python
print( len(speakeraff.Speakers[speakeraff['Aff'] == 'D']),
    len(speakeraff.Speakers[speakeraff['Aff'] == 'R']))
```
